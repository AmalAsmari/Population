```python
fig, axes = plt.subplots(2, 1,figsize=(15,12))

plt.xticks(rotation=45)
sns.barplot(data=data.head(10), x='Name', y='Density (per km²)', ax=axes[0])
sns.barplot(data=data.head(10), x='Name', y='Area (km²)', ax=axes[1])

plt.subplots_adjust(
                    bottom=0.1, 
                    top=0.9, 
                    wspace=0.4, 
                    hspace=0.4)
plt.show()
```

```python
def plottting(df):
    for i in range(len(df.index)):
        country = df.iloc[i][2:10].sort_values()
        name = df.iloc[i][1]
        growth_rate = df.iloc[i][12]
        fig = plt.figure(figsize = (10, 5))
        ax = sns.barplot(x = country.index, y = country,  palette = 'bright')
        ax.bar_label(ax.containers[0])
        plt.title(str(name)+' Population by Years')
        plt.xlabel('Years')
        plt.ylabel('Popilation')
        plt.show()

plottting(df.head(10))
```
